```python
def preprocess_data(df, dept_rename, ac_yr_rename, cgpa_rename, new_column_names, encoders, scaler, scholarship_rename, uni_rename, gender_rename, age_rename, dept_rename_2):
    df.columns = [col.strip() for col in df.columns]
    new_column_names_stripped = {k.strip(): v for k, v in new_column_names.items()}
    df.rename(columns=new_column_names_stripped, inplace=True)
    df['Academic_Year'] = df['Academic_Year'].map(ac_yr_rename)
    df['Department'] = df['Department'].map(dept_rename)
    df['CGPA'] = df['CGPA'].map(cgpa_rename)
    df['Scholarship'] = df['Scholarship'].map(scholarship_rename)
    df['University'] = df['University'].map(uni_rename)
    df['Gender'] = df['Gender'].map(gender_rename)
    df['Age'] = df['Age'].map(age_rename)
    df['Department'] = df['Department'].map(dept_rename_2)
    numerical_columns = df.select_dtypes(exclude='object').columns
    df[numerical_columns] = scaler.transform(df[numerical_columns])
    return df
```
